```python
def write_cs_resume(name, location, occuptation='None', id_num=1):
    """
    Write a generic CS resume with a specfic name, location, and one unrelated work 
    experience that is a strong he or she occupation in vector space as defined by
    'Man is to Computer Programmer as Woman is to Homemaker? Debiasing Word Embeddings' 
    https://arxiv.org/pdf/1607.06520.pdf 
    
    occuptation can take the values None, 'he', or 'she'
    """
    resume_str = ""
    if len(name.split()) == 2:
        f_name = name.split()[0]
        l_name = name.split()[1]
    else:
        f_name = l_name = name
    
    education = """University of {} | Bachelor of Science in Computer Science | 2013 - 2017
    Coursework: Analysis I and II; Introduction to Programming; Concepts of Mathematics; Principles of Imperative Programming, Great Theoretical Ideas in Computer Science""".format(location)
    
    cs_job = """Crane & Jenkins, {} | Software Engineer | March 2018 - Present""".format(location)
    
    cs_bullets = ["Integrate code changes, test code changes, and document code changes",
    "Developed Perl and PHP scripts to translate data between applications",
    "Developed automation scripts to test storage appliances in Python",
    "Practice agile development methodologies and work with current networking technologies",  
    "Used Hadoop to build a scalable distributed data solution"]
       
    if occuptation == 'he':
        occuptation = 'Barber'
    elif occuptation == 'she':
        occuptation = 'Hairdresser'
    else:
        occuptation = ""
    
    hair_job = "Top Salon, {} | {} | 2013 - 2014".format(location, occuptation)
    
    hair_bullets = ["Personally handled customer problems or concerns that ensured appropriate resolution steps were taken that maintained customer satisfaction",
                    "Built enduring client relationships and a thriving customer base of repeat and referral business.",
                    "Provided clientele with quality hair styling, coloring, and perming"]

    project = """Projects:
    
    News Delivery System (Spring 2014) - online information gathering/presentation system
    Integrated old code with new for web application delivering custom tailored web news. Coded in Java using Model-View-Controller architecture.
    
    Budgie (September 2016) - Lead Android Developer, Git Repo Manager
    App to manage and categorize expenses. Implements Microsoft’s Oxford Opitcal Character Recognition API. Pie charts show spending distribution. 
    """
    
    skills = """Skills:
    
    Java, Python, C, SML, HTML5, CSS, Django, Android, LATEX, Git, Data structures, Software design patterns """


    document = Document()
    document.add_heading(name, 0)
    resume_str += name + " "
    
    document.add_paragraph('{}.{}@gmail.com | 656-789-6599 | www.linkedin.com | github.com/{}123'.format(f_name.lower(), l_name.lower(), l_name.lower()))
    resume_str += '{}.{}@gmail.com | 656-789-6599 | www.linkedin.com | github.com/{}123'.format(f_name.lower(), l_name.lower(), l_name.lower())
    
    document.add_paragraph("Education:")
    resume_str += " Education: "
    
    document.add_paragraph(education)
    resume_str += education + " "
    
    document.add_paragraph("Work Experience:")
    resume_str += "Work Experience: "
    
    document.add_paragraph(cs_job)
    resume_str += cs_job + " "
    
    for bullet in cs_bullets:
        document.add_paragraph(bullet, style='List Bullet')
        resume_str += bullet + " "
        
    document.add_paragraph(hair_job)
    
    for bullet in hair_bullets:
        document.add_paragraph(bullet, style='List Bullet')
        resume_str += bullet + " "  
        
    document.add_paragraph(project)
    resume_str += project + " "
    
    document.add_paragraph(skills)
    resume_str += skills + " "
    
        
    document.save('{}_{}_{}_Resume.docx'.format(f_name, l_name, id_num))
    return resume_str
```

```python
test_male_names = ["Rahul Lal", "Jose Garcia", "DeShawn Washington", "Jake Johnson"]

test_female_names = ["Priya Kumar", "Isabella Rodriguez", "Imani Jackson", "Molly Smith"]

test_locations = ["San Fransisco, CA", "Mexico City, Mexico"]
```

```python
import pandas as pd 

test_male_names = ["Rahul Lal", "Jose Garcia", "DeShawn Washington", "Jake Johnson"]

test_female_names = ["Priya Kumar", "Isabella Rodriguez", "Imani Jackson", "Molly Smith"]

test_locations = ["San Fransisco, CA", "Mexico City, Mexico"]

def test_resumes(female_names:list, male_names:list, locations:list):
    names = []
    ids = []
    gender = []
    resume_location = []
    resume_strs = []
    docx_filenames = []
    text_filenames = []
    
    for i in range(len(female_names)):
        for j in range(len(locations)):
            name = female_names[i]
            location = locations[j]
            names.append(name)
            id_num = str(i)+str(j)+'f'
            ids.append(id_num)
            resume_location.append(location)
            resume_str = write_cs_resume(name, location, occuptation='she', id_num=id_num)
            resume_strs.append(resume_str)
            docx_filenames.append('{}_{}_{}_Resume.docx'.format(name.split()[0], name.split()[1], id_num))
            filename = '{}_{}_{}_Resume.txt'.format(name.split()[0], name.split()[1], id_num)
            with open('{}_{}_{}_Resume.txt'.format(name.split()[0], name.split()[1], id_num), 'w') as f:
                f.write(resume_str)
            text_filenames.append(filename) 
                
    for i in range(len(male_names)):
        for j in range(len(locations)):
            name = male_names[i]
            names.append(name)
            location = locations[j]
            resume_location.append(location)
            id_num = str(i)+str(j)+'m'
            ids.append(id_num)
            resume_str = write_cs_resume(name, location, occuptation='he', id_num=id_num)
            resume_strs.append(resume_str)
            docx_filenames.append('{}_{}_{}_Resume.docx'.format(name.split()[0], name.split()[1], id_num))
            filename = '{}_{}_{}_Resume.txt'.format(name.split()[0], name.split()[1], id_num)
            with open(filename, 'w') as f:
                f.write(resume_str)
            text_filenames.append(filename)
            
    # Resume with no name or location or gendered occupation title
    names.append("Blank")
    id_num = '001b'
    ids.append(id_num)
    gender = "None"
    resume_location.append("None")
    resume_str = write_cs_resume(name='', location='', id_num=id_num)
    resume_strs.append(resume_strs)
    docx_filenames.append('__{}_Resume.docx'.format(id_num))
    filename = 'blank_{}.txt'.format(id_num)
    with open(filename, 'w') as f:
        f.write(resume_str)
    text_filenames.append(filename)    
    
    
    resumes_df = pd.DataFrame()
    resumes_df['Name'] = names
    resumes_df['Id'] = ids
    resumes_df['Location'] = resume_location
    resumes_df['Text'] = resume_strs
    resumes_df['docx'] = docx_filenames
    resumes_df['text_file'] = text_filenames
    
    return resumes_df
```

```python
resumes_df = test_resumes(female_names=test_female_names, male_names=test_male_names, locations=test_locations)
```
